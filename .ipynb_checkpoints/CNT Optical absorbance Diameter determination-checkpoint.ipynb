```python
#Following pandas tutorial for data import and plotting: http://pandas.pydata.org/pandas-docs/stable/tutorials.html
%matplotlib inline
import pandas as pd
pd.set_option('display.mpl_style', 'default')
import numpy as np
```

```python
ni_spectrum = pd.read_csv('20150512_JH_RawA.csv', skiprows = 1) #skip that one row of BS in the file
```

```python
#make sure it looks good
ni_spectrum[:5]
```

```python
ni_spectrum.plot(x = 'Wavelength (nm)', y = 'Abs', figsize = (15,7))
```

```python
#Convert wavelength to eV
ni_spectrum['Energy (eV)'] = 4.135667516*10**(-15)*299792458/(ni_spectrum['Wavelength (nm)']*10**-9)
ni_spectrum[:5]
```

According to this: http://pubs.rsc.org.echo.louisville.edu/en/Content/ArticleLanding/2014/NR/C3NR03812B
approximate location of transitions visualized in optical absorbance spectrum: http://www.rsc.org.echo.louisville.edu/suppdata/nr/c3/c3nr03812b/c3nr03812b.pdf

S2: ~950nm (1.2 eV)
S3: ~500nm (2.5 eV)
M1: ~700nm (1.7 eV)


```python
ni_spectrum.plot(x = 'Energy (eV)', y = 'Abs', figsize = (15,7))
```

In the Nanointegris data above, looks like S2 peaks are between 0.8 and 1.2 eV

```python
bk_msspectrum = pd.read_csv('Iso M S filtrates3.csv', index_col = 'Energy (eV)')
bk_msspectrum = 2 - np.log10(bk_msspectrum)
bk_msspectrum.plot(figsize = (15, 7))
```

Spectra from Dr. Medes UV vis machine in summer/fall 2014 when I first did the metal and semiconducting experiment. Broad M1 at 1.75 eV and faint S3 peaks from 2.25 to 2.7 eV

```python
bk_msspectrum['Wavelength (nm)'] = 4.135667516*10**(-15)*299792458/(bk_msspectrum['Wavelength (nm).1']*10**-9)
```

```python
bk_volspectrum = pd.read_csv('volumeabs.csv', index_col = 'Energy (eV)')
bk_volspectrum.plot(figsize = (15, 7))
```

Spectra of the different volumes taken on Dr. O'Toole's machine. S3 peaks in the 2.25-2.7 eV range are better defined. And, I suppose that slope upward below 1.5 eV is the base of the S2 peak. A similar feature is seen in the prior graph too.
